# Objective
- Is there a problem?
- How would we solve and improve?
- How to measure whether it is improved or not?

Areas to look at:
- Diversity
- Structure
- Transparency/Communication
- Connection with peers
- Job Satisfaction
- Conflict
- Leadership
- Lifestyle
- General Feedback 

Takeaways:
In general, it may be a good idea to hire more people while keeping the diversity (minorities) in mind. People feel isolated with working from home and by using Slack, it is hard to talk to people synchronistically. It would be great if we can have more social gatherings for employees to connect with each other and employees will be able to contribute or participate in these events. 

Employees prefer the Clan Culture and wish that it could be more. 

Additionally, employees would like the relationship with their leaders to be more mentoring, facilitating, or nurturing and some reported feeling disrespectful from the higher management teams, so it may be an area to investigate more closely. They would love to have a formal mentorship program for promotion, career paths/job shadowing, conflict communication, and feedback to the company.

Hammer can also consider setting some formal rules or policies for teams to follow so that collaboration among teams feels more structured and controlled. 

Employees wish they could have more benefits e.g. regular 401k matching, fully paid health insurance for primary insured and swags (sweatshirts, coffee mugs, etc), so this could be an improvement Hammer can make if fiancially capable as I think it helps with retention and employee satisfaction.

# Responses to Questions by Categories

```
# Summary of survey by categories
data_cat_now = data_cat[data_cat.index.str.contains("NOW")].loc[:,1:]
data_cat_prefer = data_cat[data_cat.index.str.contains("PREFERRED")].loc[:,1:]
mean_cat_now = np.round_(np.mean(data_cat_now.values.astype(float),axis=1))
mean_cat_prefer = np.round_(np.mean(data_cat_prefer.values.astype(float),axis=1))
mean_cat = pd.DataFrame(np.vstack((mean_cat_now,mean_cat_prefer)).T)
mean_cat.index = data_cat[data_cat.index.str.contains("NOW")].loc[:,0]
mean_cat.columns = ["Now","Preferred"]
mean_cat["Diff"] = mean_cat.Preferred - mean_cat.Now
```

```
# less satisfied with Management Employees 
print(f'Dominant Characteristics: mean satisfaction of {mean_cat.iloc[:4].Now.mean()} with an expectation of {mean_cat.iloc[:4].Preferred.mean()}')
print(f'Organizational Leadership: mean satisfaction of {mean_cat.iloc[4:8].Now.mean()} with an expectation of {mean_cat.iloc[4:8].Preferred.mean()}')
print(f'Management Employees: mean satisfaction of {mean_cat.iloc[8:12].Now.mean()} with an expectation of {mean_cat.iloc[8:12].Preferred.mean()}')
print(f'Organizational Glue: mean satisfaction of {mean_cat.iloc[12:16].Now.mean()} with an expectation of {mean_cat.iloc[12:16].Preferred.mean()}')
print(f'Strategic Emphases: mean satisfaction of {mean_cat.iloc[16:20].Now.mean()} with an expectation of {mean_cat.iloc[16:20].Preferred.mean()}')
print(f'Criteria Success: mean satisfaction of {mean_cat.iloc[20:].Now.mean()} with an expectation of {mean_cat.iloc[20:].Preferred.mean()}')
```

```
mean_cat
```

- For Dominant Characteristics:  
    - Employees would like to have an environment that is slightly more dynamic entrepreneurial and much more controlled and structured
- For Organizational Leadership:
    - Employees would like the relationship with their leaders to be more mentoring, facilitating, or nurturing
- For Management Employees:
    - They want the management style in the organization to stress more on security of employment, conformity, predictability, and stability in relationships. Some also value individual risk-taking, innovation, freedom, and uniqueness, or hard-driving competitiveness, high demands, and achievement in terms of the management style
- For Organizational Glue:
    - They want to have more formal rules and policies to follow
- For Strategic Emphases:
    - They intend to have more permanence and stability
- For Criteria Success:
    - They hope to have successes be measured more on the basis of the development of human resources, teamwork, employee commitment, and concern for people or having the most unique or newest products
